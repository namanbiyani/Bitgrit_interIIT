```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

from sklearn.preprocessing import StandardScaler
```

```python
df_train=pd.read_csv("train.csv")
df_test=pd.read_csv("test.csv")
```

```python
df_train.head(5)
```

```python
df_test
```

```python
x_target = df_train['n_Procedure']
x_train = df_train.drop(['n_Procedure'],inplace=False,axis=1)
```

```python
x_train['Date'] = pd.to_datetime(x_train['Date'], errors='coerce')
x_train['Month'] = x_train['Date'].dt.month
x_train['Day'] = x_train['Date'].dt.day
x_train.drop(['Date'],inplace=True,axis=1)
P_1 = [1 if x is 'A' else 0 for x in x_train['Procedure']]
P_2 = [1 if x is 'B' else 0 for x in x_train['Procedure']]
P_3 = [1 if x is 'C' else 0 for x in x_train['Procedure']]
x_train['P_1'] = P_1
x_train['P_2'] = P_2
x_train['P_3'] = P_3
```

```python
x_train.drop(['Procedure'],inplace=True,axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_train, x_target, test_size=0.20, random_state=42)
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
regr = RandomForestRegressor(random_state=0)
regr.fit(X_train,y_train)
```

```python
y_pred=regr.predict(X_train)
```

```python
y_pred_roundoff=np.round(y_pred)
y_pred_roundoff
```

```python
print(np.exp(-mean_squared_error(y_train, y_pred)))
print(np.exp(-mean_squared_error(y_train, y_pred_roundoff)))
```

```python
y_pred=regr.predict(X_test)
y_pred_roundoff=np.round(y_pred)
print(np.exp(-mean_squared_error(y_train, y_pred)))
print(np.exp(-mean_squared_error(y_train, y_pred_roundoff)))
```
